```python
from googleapiclient.discovery import build
import pandas as pd
import time
import json
import os
import random
from googleapiclient.errors import HttpError

# API key from your notebook
API_KEY = "AIzaSyDyoZ13tvOSmoLKy_tPb3r06EQ89KdthQQ"

# Initialize YouTube API client
youtube = build("youtube", "v3", developerKey=API_KEY)

# File to save progress
CHECKPOINT_FILE = "youtube_scraping_checkpoint.json"

def search_videos(query, max_results=50):
    """Search for videos matching a query with retry logic."""
    video_ids = []
    next_page_token = None
    retry_count = 0
    max_retries = 5
    
    while len(video_ids) < max_results:
        try:
            request = youtube.search().list(
                part="id,snippet",
                q=query,
                type="video",
                maxResults=min(50, max_results - len(video_ids)),
                pageToken=next_page_token
            )
            response = request.execute()
            
            # Extract video IDs
            for item in response.get("items", []):
                if item["id"]["kind"] == "youtube#video":
                    video_ids.append(item["id"]["videoId"])
            
            next_page_token = response.get("nextPageToken")
            if not next_page_token:
                break
                
            # Reset retry count on successful request
            retry_count = 0
            
        except HttpError as e:
            error_details = json.loads(e.content.decode())
            reason = error_details.get("error", {}).get("errors", [{}])[0].get("reason", "")
            
            if reason == "quotaExceeded":
                print("API quota exceeded! Saving progress and exiting.")
                return video_ids
            
            if retry_count < max_retries:
                retry_count += 1
                wait_time = 2 ** retry_count + random.random()  # Exponential backoff with jitter
                print(f"YouTube API error: {e}. Retrying in {wait_time:.1f} seconds...")
                time.sleep(wait_time)
            else:
                print(f"Failed after {max_retries} retries. Moving on.")
                break
        
        except Exception as e:
            print(f"Error searching videos: {e}")
            if retry_count < max_retries:
                retry_count += 1
                wait_time = 2 ** retry_count + random.random()
                print(f"Retrying in {wait_time:.1f} seconds...")
                time.sleep(wait_time)
            else:
                break
    
    return video_ids

def get_all_comments(video_id, max_comments_per_video=None):
    """Fetch comments for a given video ID with error handling and retries."""
    comments = []
    next_page_token = None
    retry_count = 0
    max_retries = 3
    
    while True:
        # Stop if we've reached the maximum comments for this video
        if max_comments_per_video is not None and len(comments) >= max_comments_per_video:
            break
            
        try:
            request = youtube.commentThreads().list(
                part="snippet",
                videoId=video_id,
                textFormat="plainText",
                maxResults=100,  # Max per API request
                pageToken=next_page_token
            )
            response = request.execute()
            
            for item in response.get("items", []):
                # Extract comment data
                comment_snippet = item["snippet"]["topLevelComment"]["snippet"]
                
                comment_data = {
                    "video_id": video_id,
                    "comment": comment_snippet["textDisplay"],
                    "author": comment_snippet["authorDisplayName"],
                    "like_count": comment_snippet["likeCount"],
                    "published_at": comment_snippet["publishedAt"]
                }
                
                comments.append(comment_data)
                
                # Check limit after each comment
                if max_comments_per_video is not None and len(comments) >= max_comments_per_video:
                    break
            
            next_page_token = response.get("nextPageToken")
            if not next_page_token:
                break
            
            # Reset retry count on successful request
            retry_count = 0
                
        except HttpError as e:
            error_details = json.loads(e.content.decode())
            reason = error_details.get("error", {}).get("errors", [{}])[0].get("reason", "")
            
            if reason == "quotaExceeded":
                print("API quota exceeded! Returning comments collected so far.")
                return comments
            
            if retry_count < max_retries:
                retry_count += 1
                wait_time = 2 ** retry_count + random.random()
                print(f"YouTube API error: {e}. Retrying in {wait_time:.1f} seconds...")
                time.sleep(wait_time)
            else:
                print(f"Failed after {max_retries} retries. Returning comments collected so far.")
                break
            
        except Exception as e:
            print(f"Error fetching comments for video {video_id}: {e}")
            if retry_count < max_retries:
                retry_count += 1
                wait_time = 2 ** retry_count + random.random()
                print(f"Retrying in {wait_time:.1f} seconds...")
                time.sleep(wait_time)
            else:
                break
            
    return comments

def save_checkpoint(search_term, video_ids, processed_idx, all_comments):
    """Save current progress to a checkpoint file."""
    checkpoint = {
        "search_term": search_term,
        "video_ids": video_ids,
        "processed_idx": processed_idx,
        "comments_count": len(all_comments)
    }
    
    with open(CHECKPOINT_FILE, 'w') as f:
        json.dump(checkpoint, f)
    
    # Also save the current comments to CSV
    temp_df = pd.DataFrame(all_comments)
    if not temp_df.empty:
        temp_df.to_csv(f"youtube_{search_term}_comments_partial.csv", index=False)
        print(f"Saved {len(temp_df)} comments to temporary file.")

def load_checkpoint():
    """Load checkpoint if it exists."""
    if os.path.exists(CHECKPOINT_FILE):
        with open(CHECKPOINT_FILE, 'r') as f:
            return json.load(f)
    return None

# Main function to search and collect comments for a specific term
def collect_comments_for_search_term(search_term, max_videos=30, target_comments=500, resume=True):
    """
    Search for videos with the given term and collect comments
    
    Args:
        search_term: The search term to look for
        max_videos: Maximum number of videos to process
        target_comments: Target number of comments to collect
        resume: Whether to try to resume from a checkpoint
        
    Returns:
        DataFrame of collected comments
    """
    checkpoint = None
    if resume:
        checkpoint = load_checkpoint()
    
    if checkpoint and checkpoint["search_term"] == search_term:
        print(f"Resuming from checkpoint with {checkpoint['comments_count']} comments already collected")
        video_ids = checkpoint["video_ids"]
        start_idx = checkpoint["processed_idx"] + 1
        
        # Load existing comments
        partial_file = f"youtube_{search_term}_comments_partial.csv"
        if os.path.exists(partial_file):
            df_partial = pd.read_csv(partial_file)
            all_comments = df_partial.to_dict('records')
            print(f"Loaded {len(all_comments)} existing comments from partial file.")
        else:
            all_comments = []
    else:
        print(f"Searching for videos with term: '{search_term}'...")
        video_ids = search_videos(search_term, max_videos)
        print(f"Found {len(video_ids)} videos.")
        all_comments = []
        start_idx = 0
    
    processed_videos = start_idx
    
    # Save checkpoint every few videos
    checkpoint_frequency = 3
    
    for i in range(start_idx, len(video_ids)):
        video_id = video_ids[i]
        processed_videos += 1
        print(f"Fetching comments for video {i+1}/{len(video_ids)}: {video_id}...")
        
        try:
            video_comments = get_all_comments(video_id, max_comments_per_video=200)
            all_comments.extend(video_comments)
            print(f"  → Got {len(video_comments)} comments. Total so far: {len(all_comments)}")
            
            # Save checkpoint periodically
            if i % checkpoint_frequency == 0:
                save_checkpoint(search_term, video_ids, i, all_comments)
            
            # Add a small delay to avoid hitting rate limits
            time.sleep(1 + random.random())
            
            # Check if we've reached our target
            if len(all_comments) >= target_comments:
                print(f"Reached target of {target_comments} comments!")
                break
                
        except Exception as e:
            print(f"Failed to process video {video_id}: {e}")
            # Save checkpoint even on failure
            save_checkpoint(search_term, video_ids, i, all_comments)
    
    # Convert to DataFrame
    df = pd.DataFrame(all_comments)
    
    # Clear checkpoint after successful completion
    if os.path.exists(CHECKPOINT_FILE):
        os.remove(CHECKPOINT_FILE)
    
    print(f"Collection complete! Processed {processed_videos} videos and collected {len(df)} comments.")
    return df

# Run the collection for "fussball"
search_term = "soccer"
df_fussball = collect_comments_for_search_term(
    search_term=search_term,
    max_videos=500,       # Process up to 50 videos
    target_comments=1000, # Target collecting 1000 comments
    resume=True          # Try to resume from checkpoint if exists
)

# Save to CSV
output_filename = f"youtube_{search_term}_comments.csv"
df_fussball.to_csv(output_filename, index=False)

# Display sample results
print(f"\nSaved {len(df_fussball)} comments to {output_filename}")
print("\nSample comments:")
print(df_fussball.head())

# Display summary statistics for like counts
if 'like_count' in df_fussball.columns and len(df_fussball) > 0:
    print("\nLike count statistics:")
    print(f"Average likes per comment: {df_fussball['like_count'].mean():.2f}")
    print(f"Maximum likes on a comment: {df_fussball['like_count'].max()}")
    print(f"Number of comments with likes: {(df_fussball['like_count'] > 0).sum()} ({(df_fussball['like_count'] > 0).sum() / len(df_fussball) * 100:.1f}%)")
```
